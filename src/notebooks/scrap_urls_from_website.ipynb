```python
import sys
sys.path.append('../') # Retrocede una carpeta atrás
from config.paths import SRC_DIR
import os
from dotenv import load_dotenv
load_dotenv()

from utils.export_data_in_file import save_json_file

from utils.scraping_utils import ScrapingUtils
```

```python
url_to_extract_urls = "https://www.velocimotors.com.mx/"
brand_to_scrape = "Veloci"
```

```python
scraping_utils = ScrapingUtils(api_key=os.getenv("FIRECRAWL_API_KEY"))
```

### Extraer las URLs del sitio

```python
urls_tuple = scraping_utils.get_all_urls_from_website(url_to_extract_urls)
urls_tuple
```

### Procesar solamente las URLs de interés

```python
#
# ! Asegurate de cambiar a lo que quieres buscar
url_list = []
text_list_to_search = ["/motos-cfmotos/"]
texto_list_to_ignore = ["force"]
for url in urls_tuple:
    if any(text in url for text in text_list_to_search) and not any(text in url for text in texto_list_to_ignore):
        url_list.append(url)
url_list
```

### Exportar URLs a un JSON

Para no tener que volver a scrapearlas

```python
# Nombre de archivo
output_path = os.path.join(f'{SRC_DIR}/data/urls_from_websites', f'{brand_to_scrape}-urls.json')
save_json_file(urls_tuple, output_path)
print(f"Lista de URLs exportada a {output_path}")
```
