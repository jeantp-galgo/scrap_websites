## Ejemplo: Procesar m√∫ltiples productos

Si necesitas scrapear fichas t√©cnicas de varios productos:

```python
# Lista de productos a scrapear
productos = [
    {
        "url": "https://www.italika.mx/motocicleta-italika-125z-gris-34006261/p",
        "name": "Italika_125Z"
    },
    # {
    #     "url": "https://www.italika.mx/motocicleta-doble-proposito-italika-dm200-con-gps-azul-34006255-1300703478/p",
    #     "name": "Italika_DM200"
    # },
]

# Procesar todos
resultados = []
for producto in productos:
    print(f"\n{'='*60}")
    print(f"Procesando: {producto['name']}")
    print('='*60)

    try:
        result = scraper.scrape_technical_specs(
            url=producto['url'],
            product_name=producto['name']
        )
        resultados.append({
            'producto': producto['name'],
            'success': True,
            'result': result
        })
    except Exception as e:
        print(f"‚ùå Error: {e}")
        resultados.append({
            'producto': producto['name'],
            'success': False,
            'error': str(e)
        })

print(f"\n\n{'='*60}")
print(f"RESUMEN FINAL: {len([r for r in resultados if r['success']])}/{len(productos)} exitosos")
print('='*60)
```
